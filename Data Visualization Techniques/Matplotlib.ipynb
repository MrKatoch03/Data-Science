# Data Visualization Using Python






```python
# Importing Libraries 
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline  
#helps to see plot inside jupyter notebook
```

# Line Plot

```python
# X-axis
x=np.arange(1,11)
x
```

```python
# Y-axis
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Adding labels and title in the graph 
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.show()
```

```python
# Changing the attributes of Line
plt.plot(x,y,color='g',linestyle=':',linewidth=4)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.show()
```

```python
#plotting two lines in the same graph
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
# Values of x,y1,y2
x,y1,y2
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=4)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.grid(True) # To add grid in the graph
plt.show()
```

```python
# Plotting two graphs in the same row
```

```python
plt.subplot(1,2,1)   #plt.subplot(number of rows, number of column, index)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.grid(True)


plt.subplot(1,2,2) # Updated the column here 
plt.plot(x,y2,color='r',linestyle='-.',linewidth=4)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.grid(True)
plt.show()
```

```python
# Plotting two graphs in different row
```

```python
plt.subplot(2,1,1)   #plt.subplot(number of rows, number of column, index)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.grid(True)



plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=4)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(0,5,11) #equal space between two numbers
y=x**2
```

```python
#object oriented
fig=plt.figure()   #blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #[left,bottom,width,height]
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])  #Values are always between 0 and 1
axes1.plot(x,y)
axes1.set_title("Larger plot")
axes2.plot(y,x)
axes2.set_title("Smaller plot")
```

```python
# Adding a legend in the graph
fig=plt.figure()
fig=plt.figure() #height,width
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(y,x**3,label='X Cubed')
ax.legend(loc=0) # Legend location can be changed by changing the integer.
# Integer value lies between 0-10 which represents different locations for legend
```

# Bar Plot

```python
student={"Bob":87,"Matt":56,"Sam":27}
```

```python
names=list(student.keys())  # X-values
values=list(student.values()) # Y-values
```

```python
plt.bar(names,values)
plt.show()
```

```python
# Adding Title, X-label, Y-label 
plt.bar(names,values)
plt.title("Marks of students")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
# Plotting bar graph horizontally 
plt.barh(names,values,color='g')
plt.title("Marks of students")
plt.ylabel("Names")
plt.xlabel("Marks")
plt.grid(True)
plt.show()
```

# Scatter Plot
### To understand the distribution between two numerical entities


```python
x=[10,20,30,40,50,60,70,80,90] # X-Values
a=[8,1,7,2,0,3,7,3,2] # Y-values 
plt.scatter(x,a)
plt.show()
```

```python
# Changing marker type, colour and size 
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,marker='*',c='y',s=100)
plt.show()
```

```python
#Adding two markers plot in the same plot
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker='*',c='y',s=100)
plt.scatter(x,b,c='r',s=120)
plt.show()
```

```python
# Plotting two graphs in the same row
```

```python
plt.subplot(1,2,1) #plt.subplot(number of rows, number of column, index)
plt.scatter(x,a,marker='*',c='y',s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,c='r',s=120)
plt.show()
```

```python
data
```

# Pie Chart

```python
fruit=['Apple','Orange','Mango','Guava']
quantity=[67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','grey','blue','black'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%')
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
