## Importing Library 

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

## Importing Dataset

```python
df =sns.load_dataset('tips')
```

```python
df.head()
```

## Histogram 

```python
df['tip'].hist(bins=30)
```

```python
df['tip'].plot(kind='hist',bins=30)
```

```python
df['tip'].plot.hist()
```

## Area Plot 

```python
df['tip'].plot.area(alpha=0.4)
```

## Scatter Plot

```python
df.plot.scatter(x='tip',y='total_bill')
```

## Box Plot

```python
df.plot.box()
```

## Density Plot 

```python
df.plot.density()
```
