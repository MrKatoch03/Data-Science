# Data Visualization Using Python



```python
# Importing Seaborn library 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
# Importing dataset 
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

# DISTRIBUTION PLOT

```python
#For univariate feature
#density vs total_bill
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
# Most of the bill is b/w 10 and 20 dollar 
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100) 
#bin value depends upon the data
```

```python
sns.distplot(tips['total_bill'],kde=False) #kde = kernel density estimation plot
```

### Explaination of kde

```python
sns.kdeplot(tips['total_bill'])
```

```python
# Visit this for more info   https://en.wikipedia.org/wiki/Kernel_density_estimation
```

# JOINT PLOT

```python
sns.jointplot(x='total_bill',y='tip',data=tips)  #combine two distplot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="hex")  
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="reg")  
```

# PAIR PLOT

```python
sns.pairplot(tips,hue='sex',palette='coolwarm') 
#pairwise relationship for entire dataframe(mostly numerical cloumns)
# hue for categorical columns like male/female , etc
```

# RUG PLOT

```python
sns.rugplot(tips['total_bill'])
```

# CATEGORICAL PLOT

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips) 
#counts number of occurences 
```

# BOX PLOT

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

# VIOLIN PLOT

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# STRIP PLOT

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

# SWARM PLOT

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

# FACTOR PLOT

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

# REGRESSION PLOT

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
           scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```
