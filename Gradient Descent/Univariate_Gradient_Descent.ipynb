# UNIVARIATE GRADIENT DESCENT 

## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining the gradient descent function
def gradient_descent(x,y):
    m_curr=b_curr=0  # assigning slope and bias term equal to 0
    iterations=100 #number of loops 
    learning_rate = 0.002 # defining the learning rate 
    n=len(x)
    mse=[]  # mean square error
    
    for i in range(iterations):
        yp=m_curr*x+b_curr  #Linear Equation 
        cost =(1/n)*sum([val**2 for val in (y-yp)]) #MSE
        mse.append(cost)
        d_dm=-(2/n) * sum(x*(y-yp)) # derivative d cost_function/dm
        d_db=-(2/n)*sum(y-yp) # derivative d cost_function/db
        
        # Combining gradient descent and linear regression
        m_curr = m_curr - learning_rate*d_dm
        b_curr = b_curr - learning_rate*d_db
        print("m {},b {},cost {},iteration {}".format(m_curr,b_curr,cost,i))
    return(mse,iterations)

# Dataset 
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

mse,iterations=gradient_descent(x,y)

# Plotting graph between iterations vs mean squared error
iterat=np.arange(iterations)
plt.plot(iterat,mse)

plt.show()
    
```
