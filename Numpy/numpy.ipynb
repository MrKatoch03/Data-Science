## Numpy Function to Construct an Array

```python
zeros=s=np.zeros((2,5))  # Use to create an array of given dimension filled with
zeros                    # a single value 0
```

```python
rng=np.arange(15)  # Use to create a 1D array with values filled in asscending order till given range 
rng
```

```python
lspace=np.linspace(1,5,12) #np.linspace(starting_number,ending_number,total_number_of_elements)
lspace # Produces an equally spaced array between two given numbers
```

```python
emp=np.empty((4,6))
emp #Return a new array of given shape and type, with random values.
```

```python
emp_like=np.empty_like(lspace)
emp_like #used to create the new array with the same shape and type as the given array.
```

```python
ide=np.identity(45) #np.identity(total_number_of_elements)
ide # Use to produce an identity matrix
```

```python
np.ones((3,4))# Use to produce an array of given dimension filed with 1 with 
```
